```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("query.csv")

# Taking a look at dataset
print(df.info())


# Converting time into date time format
df['time'] = pd.to_datetime(df['time'], errors='coerce')

# Extracting year month and day features from datetime created above
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month
df['day_of_week'] = df['time'].dt.day_name()

```

```python
# 1. Yearly Earthquake Count**
plt.figure(figsize=(10, 5))
year_counts = df['year'].value_counts().sort_index()
plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.title('Yearly Earthquake Count')
plt.grid(True)
plt.show()
```

```python
# 2. Monthly Earthquake Count**
plt.figure(figsize=(8, 5))
month_counts = df['month'].value_counts().sort_index()
plt.bar(month_counts.index, month_counts.values, color='coral')
plt.xlabel('Month')
plt.ylabel('Number of Earthquakes')
plt.title('Monthly Earthquake Count')
plt.xticks(range(1, 13))
plt.show()
```

```python
# 3. Earthquake Count by Day of Week**
plt.figure(figsize=(8, 5))
day_counts = df['day_of_week'].value_counts()
plt.bar(day_counts.index, day_counts.values, color='seagreen')
plt.xlabel('Day of Week')
plt.ylabel('Number of Earthquakes')
plt.title('Earthquake Count by Day of Week')
plt.show()
```

```python
# 4. Magnitude Distribution**
plt.figure(figsize=(8, 5))
plt.hist(df['mag'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.title('Distribution of Earthquake Magnitudes')
plt.show()
```

```python

# 5. Magnitude vs. Depth Scatter Plot**
plt.figure(figsize=(8, 5))
plt.scatter(df['depth'], df['mag'], alpha=0.5, color='purple')
plt.xlabel('Depth (km)')
plt.ylabel('Magnitude')
plt.title('Magnitude vs. Depth of Earthquakes')
plt.show()
```
