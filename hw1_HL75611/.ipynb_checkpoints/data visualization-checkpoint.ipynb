```python
import pandas as pd

file_path = "Games.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
import matplotlib.pyplot as plt


yearly_trends = df.groupby("Year_of_Release").size()

plt.figure(figsize=(12, 6))
plt.plot(yearly_trends.index, yearly_trends.values, marker='o', linestyle='-', color='blue')
plt.title("Game Releases Over Time", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Games Released", fontsize=12)
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

top_genres = ["Action", "Sports", "Misc", "Role-Playing", "Shooter"]
df_filtered = df[df["Genre"].isin(top_genres)]

genre_trends = df_filtered.groupby(["Year_of_Release", "Genre"]).size().unstack()


plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
genre_trends.plot(kind="line", figsize=(12, 6), linewidth=2.0)

plt.title("Genre Popularity Over Time", fontsize=14)
plt.xlabel("Year of Release", fontsize=12)
plt.ylabel("Number of Games Released", fontsize=12)
plt.legend(title="Genre")

plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

regional_sales = df[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()

plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
sns.barplot(x=regional_sales.index, y=regional_sales.values, palette="Blues_r")

plt.title("Regional Sales Distribution", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Total Sales (Millions)", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

publisher_counts = df["Publisher"].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.barplot(y=publisher_counts.index, x=publisher_counts.values, color="brown")

plt.title("Top 10 Publishers by Number of Games Released", fontsize=14)
plt.xlabel("Number of Games Released", fontsize=12)
plt.ylabel("Game Publisher", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

platform_counts = df["Platform"].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.barplot(y=platform_counts.index, x=platform_counts.values, palette="Blues")

plt.title("Top 10 Gaming Platforms by Number of Releases", fontsize=14)
plt.xlabel("Number of Games Released", fontsize=12)
plt.ylabel("Gaming Platform", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.histplot(df["Global_Sales"], bins=30, kde=True, color="brown", alpha=0.5)

plt.xscale("log")

plt.title("Distribution of Global Game Sales", fontsize=14)
plt.xlabel("Global Sales (Millions)", fontsize=12)
plt.ylabel("Number of Games", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_features = ["Global_Sales", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Critic_Score", "User_Score"]
corr_matrix = df[corr_features].corr(numeric_only=True)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")

plt.title("Feature Importance Heatmap (Correlation Chart)", fontsize=14)

plt.show()
```
